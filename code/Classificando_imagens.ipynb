```python
for i in range(1,31):
  #loading image
  print('analyzing', f"brazilian{i}).jpeg")
  # Construct the full path to the image in Google Drive
  image_path = f"{caminho_arquivo}/brazilian{i}.jpeg"
  img =  cv2.imread(image_path) #loading image

  # Check if the image was loaded successfully
  if img is None:
    print(f"Error: Could not load image from {image_path}. Skipping.")
    continue

  color_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  plt.imshow( color_img)
  #this analyses the given image and gives values
  #when we use this for 1st time, it may give many errors and some google drive links to download some '.h5' and zip files, download and save them in the location where it shows that files are missing.
  prediction = DeepFace.analyze(color_img, enforce_detection = False)

  df_adobe.loc[i,'dominant_gender'] = prediction[0]['dominant_gender']
  df_adobe.loc[i,'age'] = prediction[0]['age']
  df_adobe.loc[i,'dominant_emotion'] = prediction[0]['dominant_emotion']
  df_adobe.loc[i,'dominant_race']  = prediction[0]['dominant_race']
  df_adobe.loc[i,'asian'] = prediction[0]['race']['asian']
  df_adobe.loc[i,'indian'] = prediction[0]['race']['indian']
  df_adobe.loc[i,'black']  = prediction[0]['race']['black']
  df_adobe.loc[i,'white']  = prediction[0]['race']['white']
  df_adobe.loc[i,'middle_eastern']  = prediction[0]['race']['middle eastern']
  df_adobe.loc[i,'latino_hispanic']  = prediction[0]['race']['latino hispanic']
```

```python
df_adobe['empresa'] = 'imagine'
df_adobe['indice'] = range(1, len(df_adobe) + 1)
```
